```python
# Linear Graph - A straight line with fixed slope

import matplotlib.pyplot as plt

plt.title('Basic Graph Plot')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time

import matplotlib.pyplot as plt

distance = list(range(2,11,2))
time = list(range(1,6))
print(distance)
print(time)
#d = 2 * t
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Distance Vs Time')
plt.grid()  # grid view of plot
plt.plot(time, distance, 'b d')  # syntax(x axis, yaxis, 'colour <space> shape')
plt.show()
```

```python
# Graph to plot velocity and time
# v = a * t

import matplotlib.pyplot as plt

t=list(range(1,11))
a= list(range(len(t)))
v = [a[i] * i for i in range(len(t))]

plt.plot(t, v, 'g *')
plt.grid()
plt.show()
```

```python
# number and square of numbers

n = list(range(10))
sn = [i**2 for i in n]
plt.plot(n,sn,'b o')
plt.grid()
plt.show()
```

```python
# voltage and time
import numpy as np
t = np.arange(0.1, 12.5, 0.1)  #t = [i/10 for i in range(1,95)]
v = np.sin(t) # [1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v,'g>')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-time.png')
plt.show()
```

```python
plt.pie(t,v)
plt.show()
```

```python
plt.bar(n,sn)
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013

import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
years = df.columns[2:]

incomeArkansas = df.values[3,2:]

plt.title('Income of Arkansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.grid()
plt.show()
```

```python
# (ncome of all states from 2005 to 2013)

incomeallstates = df.iloc[:,2:].mean()
years = df.columns[2:]
plt.plot(incomeallstates,'g')
plt.xlabel('Years')
plt.ylabel('Income Average')
plt.grid()
plt.show()
```

```python
# Binary Quires

n,q = input().split()
arr = list(map(int,input().split()))
for i in range(int(q)):
    qset = list(map(int,input().split()))
    if len(qset) ==2:
        arr[qset[1]-1] = arr[qset[1]-1]^1
    else:
        if (arr[qset[2]-1]&1):
            print('ODD')
        else:
            print('EVEN')
```
